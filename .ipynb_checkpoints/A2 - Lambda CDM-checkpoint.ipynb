### Part A: Prepare The Graph

We aim first to plot the components that make up the Universe, and how they evolve with the scale factor.
Starting with the axes, we would like to plot $\log_{10}{\left( \frac{\rho_x (z)}{\rho_c (z = 0)} \right)}$ vs $\log_{10}{(a)}$ – this is known as a log-log plot, and is particularly useful when you want to show power laws. 

For example, if we have a relation $X = Y^A$, it’s easier to plot on log-log axes because $\log_{10}{(X)} = A \log_{10}{(Y)}$ is a nice linear relation.

We begin this problem by making an x-axis ranging from $-4 \leq \log_{10}{a} \leq 1$ – so that we can see from a
redshift of $z \approx 10000$ through to the distant future – and a y-axis ranging from $-5 \leq \log_{10}{\left( \frac{\rho_x (z)}{\rho_c (z = 0)} \right)} \leq 20$.

```python
# Import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
# Define the factor scale for the x-axis
a_values = np.logspace(-4, 1, 1000)

# Create the plot
plt.figure(figsize = (8, 6), dpi = 100)
plt.xlabel("log10(Scale Factor)")
plt.ylabel("log10(Density)")
plt.xlim(-4, 1)
plt.ylim(-5, 15)
plt.title("Plot of the Component Density vs Scale Factor")
plt.grid(True)
plt.show()
```

### Part B: Plotting The Component Density vs Scale Factor

ΛCDM includes a few components:
* A cosmological constant, $\Lambda$, $\Omega \Lambda_o = 0.7$
* Cold Dark Matter, $\Omega C_o = 0.26$
* Baryonic Matter, $\Omega B_o = 0.04$
* Radiation, $\Omega R_o = 5 \times 10^{-5}$

Additionally, $\Omega M = \Omega C + \Omega B$ is defined as a combined matter density. Using $\Omega X = {\left( \frac{\rho_x (z)}{\rho_c (z = 0)} \right)}$ as a starting point, we plot each component’s value today and its evolution against the scale factor on the prepared axes.

First of all, we need to find how each component density evolves with the scale factor. For any component X, the parameter density X today is defined as $\Omega X_o = \frac{\rho_{Xo}}{\rho_c}$ where $\rho_{Xo}$ is the density of the object today and  $\rho_c$ is the critical density. If we know that the density evolves with the scale factor, i.e., $\rho_X (z) = \rho_{Xo} \cdot a^b$ for some constant b and $\rho_{Xo} = \Omega X_o \cdot \rho_c$ , the parameter density as a function of factor scale can be written as $\Omega X (a) = \Omega X_o \cdot a^b$.

We know that
1. Cosmological constant stays constant across factor scale $\Omega \Lambda \sim$ constant. Therefore, $\Omega \Lambda (a) = \Omega \Lambda_o$.
2. Matter density evolves inversely proportional to the cube of the scale factor $\Omega M \sim a^{-3}$. Therefore, $\Omega M (a) = \Omega M_o \cdot a^{-3}$.
3. Radiation density evolves inversely proportional to the scale factor to the power of four $\Omega M \sim a^{-4}$. Therefore, $\Omega R (a) = \Omega R_o \cdot a^{-4}$.

Next, create the plot

```python
# The density for different components
omega_A0 = 0.7   # Cosmological constant
omega_M0 = 0.3   # Combined matter density, i.e, baryonic density (0.04) + CDM density (0.26)
omega_R0 = 5e-5  # Radiation density

# Define the factor scale
a_values = np.logspace(-4, 1, 1000)

# The density of each component as a function of factor scale
omega_A = omega_A0 * np.ones_like(a_values)
omega_M = omega_M0 * a_values**(-3)
omega_R = omega_R0 * a_values**(-4)

# Create the plot
plt.figure(figsize = (8, 6), dpi = 100)
plt.plot(np.log10(a_values), np.log10(omega_A), color = 'red',         label = 'Cosmological constant')
plt.plot(np.log10(a_values), np.log10(omega_M), color = 'darkgreen',   label = 'Matter')
plt.plot(np.log10(a_values), np.log10(omega_R), color = 'orange',      label = 'Radiation')
plt.xlabel("log10(Factor Scale)")
plt.ylabel("log10(Density)")
plt.xlim(-4, 1)
plt.ylim(-5, 15)
plt.title("Plot of the Component Density vs Scale Factor")
plt.grid(True)
plt.legend(bbox_to_anchor = (1, 1), loc = "upper left")
plt.show()
```

### Part C: Dividing The Universe

From the graph, we have 3 lines, crossing one another at various points. In this section, we are to identify the crossing points and give values for the scale factor and redshift at each one. Then, we use these points to divide the history of the universe into 'eras' and identify the dominant component in each.

To find the crossing points, we only have to equate the parameter density equation of one component to another.

1. Crossing points between cosmological constant component and matter component.
\begin{gather*}
    \Omega \Lambda (a) = \Omega M (a) \\
    \Omega \Lambda_o = \Omega M_o \cdot a^{-3} \\
    a^{3} = \frac{\Omega M_o}{\Omega \Lambda_o} \\
    a = \left(\frac{\Omega M_o}{\Omega \Lambda_o}\right)^{\frac{1}{3}} = \left(\frac{0.3}{0.7}\right)^{\frac{1}{3}} \\
    a = 0.75
\end{gather*}

    Thus, in the plot, it will be located at x axis equals to $\log{(0.75)} = -0.1249$.

    To find the redshift, use the equation $\frac{1}{a} = 1 + z$,
\begin{gather*}
    \frac{1}{a} = 1 + z \\
    \frac{1}{0.75} = 1 + z \\
    z = \frac{1}{3}
\end{gather*}

2. Crossing points between cosmological constant component and radiation component.
\begin{gather*}
    \Omega \Lambda (a) = \Omega R (a) \\
    \Omega \Lambda_o = \Omega R_o \cdot a^{-4} \\
    a^{4} = \frac{\Omega R_o}{\Omega \Lambda_o} \\
    a = \left(\frac{\Omega R_o}{\Omega \Lambda_o}\right)^{\frac{1}{4}} = \left(\frac{5 \times 10^{-5}}{0.7}\right)^{\frac{1}{4}} \\
    a = 0.09
\end{gather*}

    Thus, in the plot, it will be located at x axis equals to $\log{(0.09)} = -1.0458$.

    To find the redshift, use the equation $\frac{1}{a} = 1 + z$,
\begin{gather*}
    \frac{1}{a} = 1 + z \\
    \frac{1}{0.09} = 1 + z \\
    z = 10.1
\end{gather*}

3. Crossing points between matter component and radiation component.
\begin{gather*}
    \Omega M (a) = \Omega R (a) \\
    \Omega M_o \cdot a^{-3} = \Omega R_o \cdot a^{-4} \\
    a = \frac{\Omega R_o}{\Omega M_o} = \left(\frac{5 \times 10^{-5}}{0.3}\right) \\
    a = 1.67 \times 10^{-4}
\end{gather*}

    Thus, in the plot, it will be located at x axis equals to $\log{(1.67 \times 10^{-4})} = -3.7773$.

    To find the redshift, use the equation $\frac{1}{a} = 1 + z$,
\begin{gather*}
    \frac{1}{a} = 1 + z \\
    \frac{1}{1.67 \times 10^{-4}} = 1 + z \\
    z = 5987
\end{gather*}

Using these values, we can then divide the universe into three eras.

```python
# The density for different components
omega_A0 = 0.7  # Cosmological constant
omega_M0 = 0.3  # Combined matter density, i.e, baryonic density (0.04) + CDM density (0.26)
omega_R0 = 5e-5 # Radiation density

# Define the factor scale
a_values = np.logspace(-4, 1, 1000)

# The density of each component as a function of factor scale
omega_A = omega_A0 * np.ones_like(a_values)
omega_M = omega_M0 * a_values**(-3)
omega_R = omega_R0 * a_values**(-4)

# Create the plot
plt.figure(figsize = (8, 6), dpi = 100)
plt.plot(np.log10(a_values), np.log10(omega_A), label = 'Cosmological constant', color = 'red')
plt.plot(np.log10(a_values), np.log10(omega_M), label = 'Matter',                color = 'darkgreen')
plt.plot(np.log10(a_values), np.log10(omega_R), label = 'Radiation',             color = 'orange')
plt.xlabel("log10(Factor Scale)")
plt.ylabel("log10(Density)")
plt.xlim(-4, 1)
plt.ylim(-5, 20)
plt.title("Plot of the Component Density vs Scale Factor")
plt.grid(True)

# Plot the lines dividing the universe
plt.axvline(x = -0.1249, linestyle = '--', color = 'darkblue')
plt.axvline(x = -3.7773, linestyle = '--', color = 'darkblue')
# plt.axvline(x = -1.0458, linestyle = '--', color = 'darkblue') since this line is not dividing the universe into 
# eras with a single dominating component

plt.text(-3.7773 + 0.05, 8, 'Radiation - Matter Transition', fontsize = 7, rotation = -90, rotation_mode = 'anchor')
plt.text(-0.1249 + 0.05, 7.5, 'Matter - Lambda Transition', fontsize = 7, rotation = -90, rotation_mode = 'anchor')

plt.legend(bbox_to_anchor = (1, 1), loc = "upper left")
plt.show()
```

From the plot above, we see that the universe is divided into three eras, the region on the left $\left(\log{(a)} \leq -3.7773\right)$ is dominated by radiation. Meanwhile, the region in the middle $\left(-3.7773 \leq \log{(a)} \leq -0.1249\right)$ is dominated by matter. Lastly, the region on the right $\left(\log{(a)} \geq -0.1249\right)$ is dominated by dark energy/cosmological constant.

### Part D: Calculating The Evolution of The Scale Factor

Assuming only the dominant component contributes in each era, we show that the following forms for the scale
factors $a(t)$ satisfy the Friedmann equation during Radiation, Matter, and Lambda domination: $a(t)_R = Ct^{1/2}, a(t)_M = Dt^{2/3}, a(t)_{\Lambda} = Ee^{H_ot}$. Solve for C, D, E.

During radiation domination $\left( \Omega_R = 1, \Omega_M = \Omega_{\kappa} = \Omega_{\Lambda} = 0 \right)$

Plugging in those values into the friedmann equation $\left(\frac{H}{H_o}\right)^2 = \left( \Omega_R a^{-4} + \Omega_M a^{-3} + \Omega_{\kappa} a^{-4} + \Omega_{\Lambda} \right)$, we get $\left(\frac{H}{H_o}\right)^2 = \left(a^{-4}\right)$. We know that $H = \frac{\dot a}{a}$. Hence,
\begin{gather*}
    \left(\frac{\frac{\dot a}{a}}{H_o}\right)^2 = a^{-4} \\
    \frac{\frac{\dot a}{a}}{H_o} = a^{-2} \\
    \frac{\dot{a}}{a} = H_o a^{-2} \\
    a \, da = H_o \, dt
\end{gather*}
Integrate both sides, 
\begin{gather*}
    \int a \, da = \int H_o \, dt \\
    \frac{a^2}{2} = H_ot \\
    \fbox{$a(t)_R = \left(2H_o t\right)^{\frac{1}{2}}$}
\end{gather*}

During matter domination $\left( \Omega_M = 1, \Omega_R = \Omega_{\kappa} = \Omega_{\Lambda} = 0 \right)$

Plugging in those values into the friedmann equation $\left(\frac{H}{H_o}\right)^2 = \left( \Omega_R a^{-4} + \Omega_M a^{-3} + \Omega_{\kappa} a^{-4} + \Omega_{\Lambda} \right)$, we get $\left(\frac{H}{H_o}\right)^2 = \left(a^{-3}\right)$. We know that $H = \frac{\dot a}{a}$. Hence,
\begin{gather*}
    \left(\frac{\frac{\dot a}{a}}{H_o}\right)^2 = a^{-3} \\
    \frac{\frac{\dot a}{a}}{H_o} = a^{-\frac{3}{2}} \\
    \frac{\dot{a}}{a} = H_o a^{-\frac{-3}{2}} \\
    a^{\frac{1}{2}} \, da = H_o \, dt
\end{gather*}
Integrate both sides, 
\begin{gather*}
    \int a^{\frac{1}{2}} \, da = \int H_o \, dt \\
    \frac{2}{3} a^{\frac{3}{2}} = H_ot \\
    \fbox{$a(t)_M = \left(\frac{3}{2} H_o t\right)^{\frac{2}{3}}$}
\end{gather*}

During lambda domination $\left( \Omega_{\Lambda} = 1, \Omega_R = \Omega_M = \Omega_{\kappa} = 0 \right)$

Plugging in those values into the friedmann equation $\left(\frac{H}{H_o}\right)^2 = \left( \Omega_R a^{-4} + \Omega_M a^{-3} + \Omega_{\kappa} a^{-4} + \Omega_{\Lambda} \right)$, we get $\left(\frac{H}{H_o}\right)^2 = 1$. We know that $H = \frac{\dot a}{a}$. Hence,
\begin{gather*}
    \left(\frac{\frac{\dot a}{a}}{H_o}\right)^2 = 1 \\
    \frac{\frac{\dot a}{a}}{H_o} = 1 \\
    \frac{\dot{a}}{a} = H_o \\
    \frac{1}{a} \, da = H_o \, dt
\end{gather*}
Integrate both sides, 
\begin{gather*}
    \int \frac{1}{a} \, da = \int H_o \, dt \\
    \ln{a} = H_ot \\
    \fbox{$a(t)_{\Lambda} = e^{H_ot}$}
\end{gather*}

Hence, we found $C = \left(2H_o\right)^{\frac{1}{2}}, D = \left(\frac{3}{2} H_o\right)^{\frac{2}{3}}, E = 1$.

### Part E: Plotting The Evolution of The Scale Factor

In this section, we make another plot, this time of $\log_{10}{(t)}$ vs $\log_{10}{(a(t))}$, over the same range of scale factors. Working backwards from today, we plot the evolution of the scale factor vs time, assuming a single component dominates in each era.

In part D, we showed that $a(t)_R = Ct^{1/2}, a(t)_M = Dt^{2/3}, a(t)_{\Lambda} = Ee^{H_ot}$. In this section, we are going to plot the evolution of the scale factor across time. Here, we will leave the constant in front as a prefactor that we can adjust to line up the graph and we are going to do it backwards in time.

```python
# Define some constants and prefactors that we can adjust to line up the plot
H0 = 70        # Hubble constant in units of km/s/Mpc
pref_A = 0.50  # Prefactor for lambda dominated era
pref_M = 0.99  # Prefactor for matter dominated era
pref_R = 0.22  # Prefactor for radiation dominated era

# Define the range of time for each era. 
# These values are derived from part C. Don't forget to divide it by H0 since 1/H0 gives the Hubble time
tvalues_A = np.logspace(-0.1249, 1, 100) / H0
tvalues_M = np.logspace(-3.7773, -0.1249, 100) / H0
tvalues_R = np.logspace(-10, -3.7773, 100) / H0

# Define the factor scale as a function of time, assuming a single component dominates the era
# These formulas are taken from part D
a_A = pref_A * np.exp(H0 * tvalues_A)
a_M = pref_M * (3/2 * H0 * tvalues_M)**(2/3)
a_R = pref_R * (2 * H0 * tvalues_R)**(1/2)

# Create the plot
plt.figure(figsize = (8, 6), dpi = 100)
plt.plot(np.log10(H0 * tvalues_A), np.log10(a_A), color = 'red',       label = 'Dark energy era')
plt.plot(np.log10(H0 * tvalues_M), np.log10(a_M), color = 'darkgreen', label = 'Matter era')
plt.plot(np.log10(H0 * tvalues_R), np.log10(a_R), color = 'orange',    label = 'Radiation era')

plt.text(-7.5, -3.8, '$a \propto t^{1/2}$')
plt.text(-3, -1.2, '$a \propto t^{2/3}$')
plt.text(0.1, 3.4, '$a \propto e^{kt}$')

# Add lines to separate each era and also lines to denote today
plt.axvline(x = -0.1249, linestyle = '--', color = 'darkblue')
plt.axvline(x = -3.7773, linestyle = '--', color = 'darkblue')
plt.axvline(x = 0,       linestyle = '--', color = 'darkblue')

plt.text(-3.7773 + 0.1, -3.8, '$t_{RM}$')
plt.text(-0.1249 - 0.5, -3.8, '$t_{M \Lambda}$')
plt.text(0 + 0.1, -3.8, '$t_{today}$')

plt.xlabel("log10($H_o t$)")
plt.ylabel("log10(a)")
plt.xlim(-10, 1)
plt.ylim(-6, 4)
plt.title("Plot of The Evolution of The Scale Factor")
plt.grid(True)
plt.legend(bbox_to_anchor = (1, 1), loc = "upper left")
plt.show()
```

### Part F: Calculating How The Curvature Evolves with The Scale Factor

Curvature, measured today as $\Omega_{\kappa} = 1 - \Omega_{tot}$is not included in the ΛCDM model, but is considered an extension, and also evolves with the scale factor. The curvature density $\rho_{\kappa}$ scales as $a^{-2}$.

Recent measurements show that $\Omega_{\kappa} = 0.0005 \pm 0.003$ – the present-day Universe is very flat indeed! For a moment, though, let’s assume it isn’t perfectly flat, and that the true value is $\Omega_{\kappa} = 0.0005$. 

Here, we add $\rho_{\kappa}$ to our plot of $\log_{10}{\left( \frac{\rho_x (z)}{\rho_c (z = 0)} \right)}$ vs $\log_{10}{(a)}$, and show how it evolves with the scale factor.

First, we plot the curvature.

```python
# The density for different components
omega_A0 = 0.7  # Cosmological constant
omega_M0 = 0.3  # Combined matter density, i.e, baryonic density (0.04) + CDM density (0.26)
omega_R0 = 5e-5 # Radiation density
omega_K0 = 5e-4 # Curvature

# Define the factor scale
a_values = np.logspace(-4, 1, 1000)

# The density of each component as a function of factor scale
omega_A = omega_A0 * np.ones_like(a_values)
omega_M = omega_M0 * a_values**(-3)
omega_R = omega_R0 * a_values**(-4)
omega_K = omega_K0 * a_values**(-2)

# Create the plot
plt.figure(figsize = (8, 6), dpi = 100)
plt.plot(np.log10(a_values), np.log10(omega_A), color = 'red',        label = 'Cosmological constant')
plt.plot(np.log10(a_values), np.log10(omega_M), color = 'darkgreen',  label = 'Matter')
plt.plot(np.log10(a_values), np.log10(omega_R), color = 'orange',     label = 'Radiation')
plt.plot(np.log10(a_values), np.log10(omega_K), color = 'darkblue',   label = 'Curvature')

plt.xlabel("log10(Factor Scale)")
plt.ylabel("log10(Density)")
plt.xlim(-4, 1)
plt.ylim(-5, 15)
plt.title("Plot of the Component Density vs Scale Factor")

plt.grid(True)
plt.legend(bbox_to_anchor = (1, 1), loc = "upper left")
plt.show()
```

The curvature evolves inversely proportional to the square of the scale factor. From the plot, it is evident that as time progresses, curvature becomes more significant relative to matter and radiation. However, it becomes less significant relative to lambda.
